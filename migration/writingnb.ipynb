```python
import psycopg2
from settings import DB_NAME, DB_PASSWORD, DB_USER

class Champion:
    def __init__(self):
        pass
    
    @staticmethod
    def champion_stats():
        conn = psycopg2.connect(database="games_db", user='postgres', password="afrocs221994")
        cursor = conn.cursor()
        cursor.execute("SELECT column_name FROM INFORMATION_SCHEMA.COLUMNS WHERE table_name = 'champion_stats'")
        stats = cursor.fetchall()
        cleaned_data = [item[0] for item in stats]
        return cleaned_data

    @staticmethod
    def search(champ_name):
        stats = Champion.champion_stats()
        conn = psycopg2.connect(database=DB_NAME, user=DB_USER, password=DB_PASSWORD)
        cursor = conn.cursor()
        cursor.execute(f"SELECT * FROM champion_stats WHERE name ILIKE %s", (champ_name,))
        champ = cursor.fetchone()

        if champ:
            result_dict = {stat[0]: stat[1] for stat in zip(stats, champ)}
            return result_dict
        else:
            return None

testing = Champion
print(testing.search("Aatrox"))
```
